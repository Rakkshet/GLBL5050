1.**COVID-19 Data Visualization**
Potential Source: Johns Hopkins University COVID-19 Data Repository, Our World in Data.

Bar Chart: Show the total number of COVID-19 cases in the top 10 affected countries.
Line Graph: Plot the trend of daily new cases over time for a specific country.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df_total_cases = pd.read_excel("Total Case.xlsx")
df_new_cases_india = pd.read_excel("New Cases India.xlsx")

top10_countries = df_total_cases.sort_values(by='total_cases', ascending=False).head(10)[['location', 'total_cases']]

fig, ax = plt.subplots(2, 1, figsize=(12, 10))

ax[0].bar(top10_countries.location, top10_countries.total_cases, color='red')
ax[0].set_title("COVID-19 Cases in Top 10 Affected Countries")
ax[0].set_xlabel('Country')
ax[0].set_ylabel('Total Cases')
ax[0].tick_params(axis='x', rotation=45)
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

ax[1].plot(df_new_cases_india.date, df_new_cases_india.new_cases, color='orange')
ax[1].set_title("Number of New Cases in India")
ax[1].set_xlabel('Date')
ax[1].set_ylabel('New Cases')
ax[1].tick_params(axis='x', rotation=45)
ax[1].xaxis.set_major_locator(plt.MaxNLocator(20))
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

fig.subplots_adjust(hspace=0.7)

plt.show()
```

2.**Global GDP Comparison**
Potential Source: The World Bank's World Development Indicators, International Monetary Fund (IMF).
Line Graph: Compare the GDP trends of several countries over the last decade.
Stacked Bar Chart: Display the GDP composition of a country by different sectors (agriculture, industry, services) over time.

```python
df_GDP = pd.read_excel("P5 GDP.xlsx")
df_india_sectors = pd.read_excel("India by Sectors.xlsx")

df_GDP.set_index('Year', inplace=True)

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].plot(df_GDP.China, color="red")
ax[0].plot(df_GDP.France, color="blue")
ax[0].plot(df_GDP.UK, color="black")
ax[0].plot(df_GDP.Russia, color="orange")
ax[0].plot(df_GDP.USA, color="yellow")
ax[0].legend(["China", "France", "United Kingdom", "Russia", "United States"])
ax[0].set_title("GDP Comparison in P5 Over the Years")
ax[0].set_xlabel("Years")
ax[0].set_ylabel("GDP")
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

ax[1].bar(df_india_sectors.Year, df_india_sectors['Industry'], color='red', label='Industry')
ax[1].bar(df_india_sectors.Year, df_india_sectors['Service'], bottom=df_india_sectors['Industry'], color='blue', label='Service')
ax[1].bar(df_india_sectors.Year, df_india_sectors['Agriculture'], bottom=df_india_sectors['Industry'] + df_india_sectors['Service'], color='green', label='Agriculture')

ax[1].set_title("Composition of India's GDP Over the Years")
ax[1].set_xlabel('Years')
ax[1].set_ylabel('Current USD$')
ax[1].legend()
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

fig.subplots_adjust(hspace=0.7)

plt.show()
```

3.**World Population Growth**
Potential Source: The World Bank, United Nations Population Division.
Histogram: Represent the distribution of country populations.
Scatter Plot: Illustrate the relationship between population and another variable such as land area or GDP.

```python
df_india_pop_for = pd.read_excel("India Population and Forest Area.xlsx")
df_world_pop = pd.read_excel("World Population Distribution.xlsx")

top10_pop_countries = df_world_pop.sort_values(by='Population', ascending=False).head(10)

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].bar(top10_pop_countries['Country'], top10_pop_countries['Population'], color='navy')
ax[0].set_title("Population Distribution in the Ten Most Populous Countries")
ax[0].set_xlabel("Country")
ax[0].set_ylabel("Population")
ax[0].tick_params(axis='x', rotation=45)
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

ax[1].scatter(df_india_pop_for['Population'], df_india_pop_for['Forest_Area'], color='tan')
ax[1].set_title("Scatter Plot Relation Between Population and Forest Area in India")
ax[1].set_xlabel("Population")
ax[1].set_ylabel("Forest Area (sq km)")
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

4.**Climate Change Indicators**
Potential Source: NASA's climate change dataset, Global Carbon Project.
Line Graph: Show the trend of average annual temperatures for a set of countries over the last century.
Bar Chart: Compare the total CO2 emissions by top emitting countries in the most recent year available.

```python
df_annual_temp = pd.read_excel("Annual Temperature in Canada and Saudia Arabia.xlsx")
df_co2 = pd.read_excel("CO2 Emissions in 2020.xlsx")

df_annual_temp.set_index('name', inplace=True)
top10_c02 = df_co2.sort_values(by='CO2', ascending=False).head(10)[['Country', 'CO2']]

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].plot(df_annual_temp.Canada, color="blue")
ax[0].plot(df_annual_temp.Saudi, color="red")
ax[0].legend(["Canada", "Saudi Arabia"])
ax[0].set_title("Average Annual Temperature from 1950-2100")
ax[0].set_xlabel("Years")
ax[0].set_ylabel("Average Annual Temperature")
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

ax[1].bar(top10_c02.Country, top10_c02.CO2, color='grey')
ax[1].set_title("CO2 Emissions by Top 10 Countries")
ax[1].set_xlabel('Country')
ax[1].set_ylabel('CO2 Emissions')
ax[1].tick_params(axis='x', rotation=45)
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

5.**International Trade Analysis**
Potential Source: United Nations Comtrade database, World Trade Organization.
Network Diagram (using Matplotlib extensions such as NetworkX): Visualize the trade relationships between countries.
Bar Chart: Compare the total exports and imports of the top trading countries.

```python
import networkx as nx

df_trade_direction = pd.read_excel("Trade Direction.xlsx")

G = nx.DiGraph()

for index, row in df_trade_direction.iterrows():
    G.add_edge(row['Country'], row['Counterpart'], weight=1 if row['Trade Direction'] == 'Export' else -1)

top_countries = sorted(G.in_degree, key=lambda x: x[1], reverse=True)[:20]
top_countries_names = [country for country, _ in top_countries]

subgraph = G.subgraph(top_countries_names)

pos = nx.spring_layout(subgraph)
nx.draw(subgraph, pos, with_labels=True, node_size=700, font_size=8, node_color='teal', font_color='black', font_weight='bold', edge_color='gray', linewidths=0.5, arrowsize=10)

plt.title("Trade Direction Network of Top 20 Countries")
plt.show()
```

```python
df_net_export = pd.read_excel("Net Export.xlsx")

top10_net_export = df_net_export.sort_values(by='NET EXPORT', ascending=False).head(10)[['Country Name', 'NET EXPORT']]

fig, ax = plt.subplots(figsize=(10, 8))

ax.bar(top10_net_export["Country Name"], top10_net_export["NET EXPORT"], color='coral')
ax.set_title("Top 10 Countries Based on Net Export")
ax.set_xlabel('Country')
ax.set_ylabel('Net Export')
ax.tick_params(axis='x', rotation=45)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

7.**Educational Attainment**
Potential Source: UNESCO Institute for Statistics, World Bank Education Statistics.
Bar Chart: Display literacy rates for different age groups within a country.
Box Plot: Show the distribution of average years of schooling across continents.

```python
df_lit_rate = pd.read_excel("Literacy Rates.xlsx")
df_avg_sch = pd.read_excel("Average Years of Schooling.xlsx")

fig, ax = plt.subplots(2, 1, figsize=(10, 10))

togo_data = df_avg_sch[df_avg_sch['Country'] == 'Togo']['Mean']
ax[0].boxplot(togo_data, vert=False, patch_artist=True)
ax[0].set_title('Average Years of Schooling in Togo')
ax[0].set_xlabel('Years of Schooling')

ax[1].bar(df_lit_rate.Country, df_lit_rate['Youth'], color='red', label='Youth literacy rate (15-24 years)')
ax[1].bar(df_lit_rate.Country, df_lit_rate['Elderly'], bottom=df_lit_rate['Youth'], color='blue', label='Elderly literacy rate (65+ years)')
ax[1].set_title("Literacy Rate for Different Age Groups")
ax[1].set_xlabel('Country')
ax[1].set_ylabel('Percentage')
ax[1].tick_params(axis='x', rotation=45)
ax[1].legend()
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

8.**Renewable Energy Usage**
Potential Source: International Energy Agency, The World Bank.
Pie Chart: Illustrate the proportion of different types of renewable energy sources in a country's energy mix.
Stacked Area Chart: Show the change in renewable versus non-renewable energy consumption over time in a country.

```python
df_energy_india = pd.read_excel("India Energy Distribution.xlsx")
df_ren_india = pd.read_excel("Renewable and Non-Renewable Energy.xlsx")

df_energy_india = df_energy_india[df_energy_india['Country'] == 'India']
df_ren_india.set_index('Year', inplace=True)

energy_sources = df_energy_india.columns[1:]
energy_values = df_energy_india.iloc[0, 1:]

fig, ax = plt.subplots(2, 1, figsize=(10, 10))

wedges, texts, autotexts = ax[0].pie(energy_values, labels=None, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
ax[0].set_title("Energy Distribution in India")
ax[0].legend(wedges, energy_sources, title="Energy Sources", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
ax[0].axis('equal')

ax[1].stackplot(df_ren_india.index, df_ren_india['Renewable'], df_ren_india['Non-Renewable'], labels=['Renewable', 'Non-Renewable'], colors=['green', 'gray'])
ax[1].set_title("Distribution between Renewable and Non-Renewable Energy in India")
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Percentage')
ax[1].tick_params(axis='x', rotation=45)
ax[1].legend()
ax[1].grid(axis='y', linestyle='--', alpha=0.7)
ax[1].xaxis.set_major_locator(plt.MaxNLocator(20))

plt.tight_layout()
plt.show()
```

10.**Life Expectancy and Health Factors**
Potential Source: Global Health Observatory data from the World Health Organization, Human Development Reports from the United Nations Development Programme.
Scatter Plot: Show the correlation between life expectancy and a health factor like BMI or smoking rates.
Violin Plot: Display the distribution of life expectancy across different regions.

```python
import seaborn as sns

df_life_BMI = pd.read_excel("Life and BMI.xlsx")
df_life = pd.read_excel("Life Expectency.xlsx")

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].scatter(df_life_BMI['Life'], df_life_BMI['BMI'], color='purple')
ax[0].set_title("Scatter Plot Relation Between Life Expectency and BMI")
ax[0].set_xlabel("Life Expectency")
ax[0].set_ylabel("BMI")
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

sns.violinplot(x='Continent', y='Life', data=df_life, ax=ax[1], palette='muted')
ax[1].set_title("Distribution of Life Expectancy Across Different Regions")
ax[1].set_xlabel("Continent")
ax[1].set_ylabel("Life Expectancy")
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
